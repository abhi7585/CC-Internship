```python
# Library for Data Manupulation
import pandas as pd

# Libraries for Data Visualization
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="darkgrid")


# Library for missing data 
import missingno as msno
```

# Importing the Data

```python
df = pd.read_csv("MumbaiCollege.csv")
```

```python
df.head()
```

# Columns

```python
df.columns
```

# Questions asked based on the data and its answers in the form of numerics and graphs

## Total number of colleges?

```python
len(df)
```

## Total number of College's in each District?

```python
df['District'].value_counts()
```

```python
plt.figure(figsize = (12,6))
sns.countplot(df['District'])
```

## Different types of status?

```python
df['Status'].value_counts()
```

```python
plt.figure(figsize = (15,8))
sns.countplot(y =df['Status'])
```

```python
plt.figure(figsize = (10, 6))
sns.countplot(y = df[df['Status'] != 'Un-Aided']['Status'])
```

## Total number of Autonomous Colleges?

```python
df['Autonomy_Status'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(df['Autonomy_Status'])
```

## What are the different minority status?

```python
df['Minority_Status'].value_counts()
```

```python
plt.figure(figsize = (10,8))
sns.countplot(y = df[df['Minority_Status'] != 'Non-Minority']['Minority_Status'])
```

## What are different years in which colleges are established?

```python
df['Establishment'].value_counts()
```

```python
plt.figure(figsize= (15, 13))
sns.countplot( y = df['Establishment'])
```

## Oldest College is ?

```python
df[df['Establishment'] == df['Establishment'].min()]['College_Name']
```

## Newest College is ?

```python
df[df['Establishment'] == df['Establishment'].max()]['College_Name']
```

## Different types of Courses available in the Mumbai region?

```python
df['Course'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(df['Course'])
```
